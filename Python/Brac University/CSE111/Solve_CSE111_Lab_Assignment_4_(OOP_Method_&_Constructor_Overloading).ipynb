### Question 1

Suppose your little sibling wants your help to check his math homework. He is done with his homework but wants you to see if all his results are correct. Since the student with all correct results gets 3 stars. However, you want your brother to check this on his own. So, you design a calculator for him in python. You could have given your scientific calculator but you wanted to give him a basic calculator and also wanted to see if you can even design one.

Subtasks:
1. Create a class called Calculator.
2. Your class shall have 1 constructor and 4 methods, namely add, subtract, multiply and divide.
3. Now, create an object of your class. After creating an object, it should print “Let’s Calculate!”
4. Then take 3 inputs from the user: first value, operator, second value
5. Now based on the given operator, call the required method and print the result.

###Sample Input
1<br>
+<br>
2<br>

###Sample Output:
Let’s Calculate!<br>
Value 1: 1<br>
Operator: +<br>
Value 2: 2<br>
Result: 3<br>

```python
class Calculator:

  def __init__(self):
    print("Let's Calculate")

  def add(self, value_1, value_2):
    print("Result: {}".format(value_1 + value_2))

  def subtract(self, value_1, value_2):
    print("Result: {}".format(value_1 - value_2))
  
  def multiply(self, value_1, value_2):
    print("Result: {}".format(value_1 * value_2))

  def divide(self, value_1, value_2):
    print("Result: {}".format(value_1 / value_2))

calc = Calculator()

value_1 = int(input())
print("Value 1:", value_1)
operator = input()
print("Operator:", operator)
value_2 = int(input())
print("Value 2:", value_2)

if operator == '+':
  calc.add(value_1, value_2)
elif operator == '-':
  calc.subtract(value_1, value_2)
elif operator == '*':
  calc.multiply(value_1, value_2)
elif operator == '/':
  calc.divide(value_1, value_2)  

```

```python
class Customer:

  def __init__(self, name):
    self.name = name

  def greet(self, name = None):
    if name == None:
      print('Hello!')
    else:
      print(f'Hello! {name}')

  def purchase(self, *items):
    print(f'{self.name}, you purchased {len(items)} item(s):')
    for item in items:
      print(item)


customer_1 = Customer("Sam")
customer_1.greet()
customer_1.purchase("chips", "chocolate", "orange juice")
print("-----------------------------")
customer_2 = Customer("David")
customer_2.greet("David")
customer_2.purchase("orange juice")
```

```python
class Vehicle:

  def __init__(self):
    self.x = 0
    self.y = 0

  def print_position(self):
    print(f"({self.x}, {self.y})")

  def moveUp(self):
    self.y += 1
  
  def moveLeft(self):
    self.x -= 1
  
  def moveDown(self):
    self.y -= 1

  def moveRight(self):
    self.x += 1

car = Vehicle()
car.print_position()
car.moveUp()
car.print_position()
car.moveLeft()
car.print_position()
car.moveDown()
car.print_position()
car.moveRight()
```

```python
class Vaccine:
  
  def __init__(self, vaccine_name, country, duration_after_first_dose):
    self.vaccine_name = vaccine_name
    self.country = country
    self.duration_after_first_dose = duration_after_first_dose

class Person:
  
  def __init__(self, person_name, *info):
    self.person_name = person_name

    if len(info) == 1:
      self.age = info[0]
      self.occupation = "General Citizen"
    elif len(info) == 2:
      self.age = info[0]
      self.occupation = info[1]

    self.vaccine_dose = 0
    self.vaccine = []

    if self.occupation != "Student" and self.age < 25:
      self.permission = "not permissible"
    else:
      self.permission = "permission"
  

  def pushVaccine(self, *vaccine_info):

    if self.permission == "not permissible":
      print(f"Sorry {self.person_name}, Minimum age for taking vaccines is 25 years now.")
      return None


    if(self.vaccine != []):
      if self.vaccine.vaccine_name != vaccine_info[0].vaccine_name:
        print(f"Sorry {self.person_name}, you can’t take 2 different vaccines")
        return None

    if len(vaccine_info) == 1:
      self.vaccine = vaccine_info[0]
    elif len(vaccine_info) == 2:
      self.vaccine = vaccine_info[0]
      self.dose_status = vaccine_info[1]
    self.vaccine_dose += 1

    if self.vaccine_dose == 1:
      print(f"1st dose done for {self.person_name}")
    elif self.vaccine_dose == 2:
      print(f"2nd dose done for {self.person_name}")
  

  def showDetail(self):
    print(f"Name: {self.person_name} Age: {self.age} Type: {self.occupation}\nVaccine name: {self.vaccine.vaccine_name}")
    if self.vaccine_dose == 1:
      print("1st dose: Given")
      print(f"2nd dose: Please come after {self.vaccine.duration_after_first_dose} days")
    elif self.vaccine_dose == 2:
      print("1st dose: Given")
      print("2nd dose: Given")





astra = Vaccine("AstraZeneca", "UK", 60)
modr = Vaccine("Moderna", "UK", 30)
sin = Vaccine("Sinopharm", "China", 30)
p1 = Person("Bob", 21, "Student")
print("=================================")
p1.pushVaccine(astra)
print("=================================")
p1.showDetail()
print("=================================")
p1.pushVaccine(sin, "2nd Dose")
print("=================================")
p1.pushVaccine(astra, "2nd Dose")
print("=================================")
p1.showDetail()
print("=================================")
p2 = Person("Carol", 23, "Actor")
print("=================================")
p2.pushVaccine(sin)
print("=================================")
p3 = Person("David", 34)
print("=================================")
p3.pushVaccine(modr)
print("=================================")
p3.showDetail()
print("=================================")
p3.pushVaccine(modr, "2nd Dose")
```
