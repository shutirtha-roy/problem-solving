```python
import math

class Cylinder:

  radius = 5
  height = 18

  def __init__(self, radius, height):
    print(f"Default radius={Cylinder.radius} and height={Cylinder.height}.")
    Cylinder.radius = radius
    Cylinder.height = height
    print(f"Updated: radius={Cylinder.radius} and height={Cylinder.height}.")
  
  @staticmethod
  def area(radius, height):
    print(f"Area: {2 * math.pi * (radius ** 2) + 2 * math.pi * radius * height}")

  @staticmethod
  def volume(radius, height):
    print(f"Volume: {math.pi * (radius**2) * height}")

  @classmethod
  def swap(cls, radius, height):
    return cls(height, radius)

  @classmethod
  def changeFormat(cls, info):
    new_radius, new_height = info.split("-")
    return cls(float(new_radius), float(new_height))

  


c1 = Cylinder(0,0)
Cylinder.area(c1.radius,c1.height)
Cylinder.volume(c1.radius,c1.height)
print("===============================")
c2 = Cylinder.swap(8,3)
c2.area(c2.radius,c2.height)
c2.volume(c2.radius,c2.height)
print("===============================")
c3 = Cylinder.changeFormat("7-13")
c3.area(c3.radius,c3.height)
c3.volume(c3.radius,c3.height)
print("===============================")
Cylinder(0.3,5.56).area(Cylinder.radius,Cylinder.height)
print("===============================")
Cylinder(3,5).volume(Cylinder.radius,Cylinder.height)
```

```python
class SultansDine:

  total_branch = 0
  total_sell = 0
  branch_info = []

  def __init__(self, location):
    self.__location = location
    self.__branch_sell = 0
    SultansDine.total_branch += 1

  @classmethod
  def details(cls):
    print(f"Total Number of branch(s): {cls.total_branch}\nTotal Sell: {cls.total_sell} Taka")
    for info in SultansDine.branch_info:
      print(f"Branch Name: {info[0]}, Branch Sell: {info[1]} Taka")
      branch_sell_percentage = (info[1] / cls.total_sell) * 100
      print(f"Branch consists of total sell's: {'%.2f'%branch_sell_percentage}%")

  def sellQuantity(self, sell_quantity):
    if sell_quantity < 10:
      self.__branch_sell = sell_quantity * 300
    elif sell_quantity < 20:
      self.__branch_sell = sell_quantity * 350
    else:
      self.__branch_sell = sell_quantity * 400
    
    SultansDine.total_sell += self.__branch_sell

  def branchInformation(self):
    print(f"Branch Name: {self.__location}\nBranch Sell: {self.__branch_sell} Taka")
    SultansDine.branch_info.append((self.__location, self.__branch_sell))


SultansDine.details()
print('########################')
dhanmodi = SultansDine('Dhanmondi')
dhanmodi.sellQuantity(25)
dhanmodi.branchInformation()
print('-----------------------------------------')
SultansDine.details()
print('========================')
baily_road = SultansDine('Baily Road')
baily_road.sellQuantity(15)
baily_road.branchInformation()
print('-----------------------------------------')
SultansDine.details()
print('========================')
gulshan = SultansDine('Gulshan')
gulshan.sellQuantity(9)
gulshan.branchInformation()
print('-----------------------------------------')
SultansDine.details()
```
